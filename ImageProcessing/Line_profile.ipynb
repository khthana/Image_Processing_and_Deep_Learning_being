```
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
import matplotlib.pylab as plt
import numpy as np
```

##Read Image

```
myImg = cv.imread('can.png')
gray_img = cv.cvtColor(myImg, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```
x, y = 120, 150
color = (0, 255, 0)
markerType = cv.MARKER_CROSS
markerSize = 15
thickness = 2
cv.drawMarker(myImg, (x, y), color, markerType, markerSize, thickness)

cv2_imshow(myImg)
```

# Line Profile

```
from matplotlib.pyplot import ylim
line = gray_img[150,:]
ylim([0, 255])
plt.plot(line)
```

# Gradient

```
from matplotlib.pyplot import ylim
line = gray_img[150,:]
line1 = np.array([*line, 0])
line2 = np.array([0, *line])
grad = line1-line2
grad[0] = 0
ylim([0, 255])
plt.plot(line)
plt.plot(grad)
```
