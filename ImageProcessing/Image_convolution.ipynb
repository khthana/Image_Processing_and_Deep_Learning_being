```
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
import matplotlib.pylab as plt
import numpy as np
```

##Read Image

```
myImg = cv.imread('dog.jpeg')
cv2_imshow(myImg)
```

```
#edge detection filter
kernel_edge_x = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernel_edge_y = kernel_edge_x.T
edge_x = cv.filter2D(myImg,-1,kernel_edge_x)
edge_y = cv.filter2D(myImg,-1,kernel_edge_y)
#Display images
cv2_imshow(myImg)
cv2_imshow(edge_x)
cv2_imshow(edge_y)
```
